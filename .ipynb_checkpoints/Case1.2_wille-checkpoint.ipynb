```python

# calling required libraries

import numpy as np
import time
import warnings
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import normalize
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
from keras.utils import to_categorical
from keras import models, layers
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
clasifier = Sequential()
```

```python
data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
df = pd.read_csv(data, sep=',' , header = None)
df.columns = ["age", "sex", "cp", "treshbps", "chol", "fbs", "restecg", "thalach",
              "exang", "oldpeak", "slope", "ca", "thal", "num"]

#df.rename_axis('.').rename_axis("ID", axis="columns")

df.index.name = 'id'

# Fill missing data with columnwise median values
df = df.fillna(df.median())

# Randomize rows
df = shuffle(df)
df.tail()

# Scale the data
#df_min = df.min()
#df_max = df.max()
#df_norm = (df - df_min)/(df_max - df_min)
#DOESENT WORK!

#df_min = df.min()
#df_max = df.max()
#df_norm = (df - df_min)/(df_max - df_min)
df.head()
```

```python
# We will make use of ScikitLearn’s ‘train_test_split’ function to divide our data 
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Starting to built NN
# Initializing Neural Network

#Our first parameter is output_dim. It is simply the number of nodes you want to add to this layer. 
#init is the initialization of Stochastic Gradient Decent. 
#In Neural Network we need to assign weights to each mode which is nothing but importance of that node. 
#At the time of initialization, weights should be close to 0 and we will randomly initialize weights using uniform function. 
#input_dim parameter is needed only for first layer as model doesn’t know the number of our input variables. 
#Remember in our case, the total number of input variables are 11. 
#In the second layer model automatically knows the number of input variable from the first hidden layer.

# Adding the input layer and the first hidden layer
#clasifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))            THESE ARE GIVING ERRORS
clasifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
#clasifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))                             THESE ARE GIVING ERRORS
clasifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
#clasifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))                           THESE ARE GIVING ERRORS
clasifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid' ))
```

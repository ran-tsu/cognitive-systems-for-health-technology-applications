```python

# calling required libraries

import numpy as np
import time
import warnings
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import normalize
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
from keras.utils import to_categorical
from keras import models, layers
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
clasifier = Sequential()
```

```python
data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
df = pd.read_csv(data, sep=',' , header = None)
df.columns = ["age", "sex", "cp", "treshbps", "chol", "fbs", "restecg", "thalach",
              "exang", "oldpeak", "slope", "ca", "thal", "num"]

#df.rename_axis('.').rename_axis("ID", axis="columns")

df.index.name = 'id'

# Fill missing data with columnwise median values
df = df.fillna(df.median())

# Randomize rows
df = shuffle(df)
df.tail()

# Scale the data
#df_min = df.min()
#df_max = df.max()
#df_norm = (df - df_min)/(df_max - df_min)
#DOESENT WORK!

#df_min = df.min()
#df_max = df.max()
#df_norm = (df - df_min)/(df_max - df_min)
df.head()
```

```python
# Create matrix of features and matrix of target variable
x = df.iloc[:, 0:13].values
y = df.iloc[:, 13].values
x
```

```python
# We will make use of ScikitLearn’s ‘train_test_split’ function to divide our data 
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Starting to built NN
# Initializing Neural Network

# Adding the input layer and the first hidden layer
clasifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
clasifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))

# Adding the output layer
clasifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

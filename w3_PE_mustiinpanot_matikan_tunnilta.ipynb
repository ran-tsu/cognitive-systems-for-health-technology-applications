```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 0], [1, 1], [1, 2]])
y = np.array([[4, 7, 9]]).T

M = la.inv(X.T.dot(X)).dot(X.T).dot(y)
print(M)
```

```python
x = np.array([[0, 1, 2]]).T
plt.plot(x, y, 'o')

xx = np.linspace(0, 2, 100).reshape(100,1)
XX = np.concatenate((np.ones((100,1)), xx), axis=1)

yy = XX.dot(M)
#print (XX)
plt.plot(xx, yy)
plt.show()
```

```python
xx = np.linspace(0, 2, 100).reshape(100,1)
yy = M[1,0] + M[0,0]*xx
plt.figure(2)
plt.plot(xx, yy)
plt.show()
```

```python
x = np.arange(1, 11)
y = 100*(x**3) - 16.2*(x**2) + 3.3*x + 0.5 + 10*np.random.randn(10)

x = x.reshape(10,1)
y = y.reshape(10,1)

X = np.concatenate((np.ones((10,1)),x, x**2, x**3),axis=1)
print(X)
print(" ")
print(y)
print(" ")

M = la.inv(X.T @ X) @ X.T @ y
print(M)

h = X @ M

plt.plot(x, y, '.')
plt.plot(x, h)
plt.show()
```
